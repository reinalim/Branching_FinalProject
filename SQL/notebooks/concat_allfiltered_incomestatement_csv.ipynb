## Grabbing all income_statement CSV's to put into 1 file.

```python
# Importing dependencies.
from glob import glob
import pandas as pd
```

```python
# Creating a list of all the balance_sheet files.
income_statement_files = sorted(glob('raw_dataset/filtered_income_statement/*filtered_income_statement.csv'))
income_statement_files
```

```python
# Create a dataframe to house all concatenated files.
df = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in income_statement_files), ignore_index = True)
df
```

```python
# Dropping unwanted columns.
cleaned_df = df.drop(['filename'],axis=1)
cleaned_df
```

```python
# Count of missing values of the entire dataframe. If any drop.
cleaned_df.isnull().sum().sum()
```

```python
# Replacing the None values with 0.
filtered_income_statement_df = cleaned_df.replace('None', 0)
filtered_income_statement_df
```

```python
# Renaming column names.

filtered_income_statement_df = filtered_income_statement_df.rename(columns = {
    'SYMBOL': 'symbol', 
    'fiscalDateEnding': 'fiscaldate_end', 
    'reportedCurrency': 'reported_currency',
    'grossProfit': 'gross_profit',
    'totalRevenue': 'total_revenue', 
    'costOfRevenue': 'costof_revenue', 
    'costofGoodsAndServicesSold': 'costof_goodservices_sold',
    'operatingIncome': 'operating_income',
    'sellingGeneralAndAdministrative': 'selling_gen_admin', 
    'researchAndDevelopment': 'research_dev', 
    'operatingExpenses': 'operating_expenses',
    'investmentIncomeNet': 'investment_income_net',
    'interestIncome': 'interest_income', 
    'interestExpense': 'interest_expense', 
    'nonInterestIncome': 'noninterest_income',
    'otherNonOperatingIncome': 'othernon_operatingincome',
    'depreciation': 'depreciation', 
    'depreciationAndAmortization': 'dpr_and_amort', 
    'incomeBeforeTax': 'income_beforetax',
    'incomeTaxExpense': 'incometax_expense',
    'interestAndDebtExpense': 'interest_debt_expense', 
    'netIncomeFromContinuingOperations': 'netincome_cont_operations', 
    'comprehensiveIncomeNetOfTax': 'comprehensive_income_netoftax',
    'ebit': 'ebit',
    'ebitda': 'ebitda', 
    'netIncome': 'netincome',
    })
filtered_income_statement_df
```

```python
# Exporting balance_sheet dataframe to csv file without the index.
filtered_income_statement_df.to_csv('raw_dataset/filtered_income_statement.csv', index = False)
```
